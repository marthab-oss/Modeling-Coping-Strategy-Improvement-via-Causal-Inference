## Dataset Overview

[Provide a high-level overview of the dataset. This should include the source of the dataset, the number of samples, the number of features, and example showing the structure of the dataset.]


```python
# Load the data
# Replace 'your_dataset.csv' with the path to your actual dataset
df = pd.read_csv('/Users/libraryfree/Desktop/MachineLearning/CausalInf_project/sequential_mediation_dataset.csv')

# Number of samples
num_samples = df.shape[0]

# Number of features
num_features = df.shape[1]

# Display these dataset characteristics
print(f"Number of samples: {num_samples}")
print(f"Number of features: {num_features}")
#print(df.head().to_markdown(index=False))

# Display the first few rows of the dataframe to show the structure
rows=8
fig, ax = plt.subplots(figsize=(18, 2))
ax.axis('off')
table = ax.table(
    cellText=df.head(rows).round(3).values,
    colLabels=df.columns,
    loc='center'
)
ax.set_title('Dataset excerpt', fontsize=14, fontweight='bold', pad=15)
table.auto_set_font_size(False)
table.set_fontsize(10)
plt.tight_layout()
plt.savefig('/Users/libraryfree/Desktop/MachineLearning/CausalInf_project/dataset_preview.jpeg', bbox_inches='tight', dpi=300)
#plt.show()
```

## Correlations

[Explore correlations between features and the target variable, as well as among features themselves.]


```python
# Example: Plotting a heatmap to show feature correlations
plt.figure(figsize=(11,8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='YlOrBr')
plt.xticks(rotation=15)
plt.yticks(rotation=20)
plt.tight_layout()
#plt.savefig('/Users/libraryfree/Desktop/MachineLearning/CausalInf_project/Correlation_Matrix.jpeg')
plt.show()
```
